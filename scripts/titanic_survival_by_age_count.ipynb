```python
"""
هیستوگرام تعداد افراد بر اساس سن، تفکیک‌شده بین بازمانده‌ها و قربانی‌ها
"""

import seaborn as sns
import matplotlib.pyplot as plt
import arabic_reshaper
from bidi.algorithm import get_display

# تابع اصلاح متن فارسی
def fix_farsi(text):
    return get_display(arabic_reshaper.reshape(text))

# دیتاست آماده تایتانیک
df = sns.load_dataset("titanic")

# رسم نمودار
plt.figure(figsize=(8,5))
sns.histplot(data=df, x="age", hue="survived", multiple="stack", bins=20)

plt.title(fix_farsi("تعداد بازمانده‌ها و قربانی‌ها بر اساس سن"), fontsize=14)
plt.xlabel(fix_farsi("سن"))
plt.ylabel(fix_farsi("تعداد"))
plt.grid(axis="y", linestyle="--", alpha=0.7)

# ذخیره خروجی
plt.savefig("../outputs/titanic_age_count.png", dpi=300, bbox_inches="tight")
plt.show()
```
